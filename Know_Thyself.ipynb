```python
import re

# Polish to English month mapping
polish_months = {
    "stycznia": "January", "lutego": "February", "marca": "March",
    "kwietnia": "April", "maja": "May", "czerwca": "June",
    "lipca": "July", "sierpnia": "August", "września": "September",
    "października": "October", "listopada": "November", "grudnia": "December"
}

def clean_date(date_str):
    """ Cleans and converts Polish date strings to a parsable English format. """
    if pd.isna(date_str):  # Handle missing values
        return np.nan
    date_str = re.sub(r"^\w+, ", "", date_str)  # Remove day names (e.g., "piątek, ")
    for pl, en in polish_months.items():
        date_str = date_str.replace(pl, en)  # Replace Polish months with English
    return date_str

# Apply cleaning
df['created'] = df['created'].apply(clean_date)
df['modified'] = df['modified'].apply(clean_date)

# Convert to datetime format after cleaning
df['created'] = pd.to_datetime(df['created'], format="%d %B %Y, %H:%M", errors='coerce')
df['modified'] = pd.to_datetime(df['modified'], format="%d %B %Y, %H:%M", errors='coerce')

# Check the results
print(df[['created', 'modified']].head(10))
df.info()  # Confirm datetime columns are fixed
```

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords

# ✅ Download stopwords if not available
nltk.download('stopwords')

############################
# 3) TOKENIZATION
############################

# If you want a quick tokenization approach:
def simple_tokenizer(text):
    if pd.isna(text):  # Handle missing values
        return []
    # Lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(f"[{string.punctuation}]", " ", text)
    # Split on whitespace
    tokens = text.split()
    return tokens

# Confirm the correct text column
print(df.columns)  # Debugging step to check available columns

# Use the correct column name (update if necessary)
text_column = 'message'  # Change this to 'content' or the actual column name if needed

# Convert each row’s text into a list of tokens
df['tokens'] = df[text_column].apply(simple_tokenizer)

# Flatten everything into a long DataFrame of [word, post_id, etc.]
long_tokens = []
for i, row in df.iterrows():
    for w in row['tokens']:
        long_tokens.append({
            'id': row.get('id', i),   # or discussion id
            'word': w,
            'rating': row.get('rating_1_10', None),  # Update column name if necessary
        })

token_df = pd.DataFrame(long_tokens)
print(token_df.head(10))

############################
# 4) REMOVE STOPWORDS ETC.
############################

# ✅ Ensure stopwords are available
stop_words_en = set(stopwords.words('english'))  # English only

# Filter out numeric-only tokens or <3 length words, plus English stopwords
def keep_token(t):
    # e.g., remove tokens that are purely numeric
    if t.isdigit():
        return False
    # remove short tokens
    if len(t) < 3:
        return False
    # remove stopwords
    if t in stop_words_en:
        return False
    return True

token_df = token_df[token_df['word'].apply(keep_token)]
print("After cleaning:", token_df.shape)
```

```python
import pandas as pd
from textblob import TextBlob

##########################
# 2) TEXTBLOB SENTIMENT
##########################

def get_textblob_sentiment(message):
    if pd.isna(message):  # Handle missing values
        return None
    blob = TextBlob(message)
    return blob.sentiment.polarity  # range: -1.0 (negative) to +1.0 (positive)

# Apply sentiment analysis
df['textblob_score'] = df['message'].apply(get_textblob_sentiment)

# Quick check of new column:
print(df[['id','textblob_score']].head(10))
```

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# ✅ Download VADER lexicon if not available
nltk.download('vader_lexicon')

############################
# 3) NLTK VADER SENTIMENT
############################
sia = SentimentIntensityAnalyzer()

def get_vader_score(text):
    if pd.isna(text):  # Handle missing values
        return {'compound': 0}  # Neutral score for NaN
    return sia.polarity_scores(text)

# Ensure the correct text column is used
text_column = 'message'  # Change to 'content' if necessary

df['vader_dict'] = df[text_column].apply(get_vader_score)

# Extract compound score only (for a single sentiment measure)
df['vader_compound'] = df['vader_dict'].apply(lambda d: d['compound'])

# Show the first few results
print(df[['id', 'vader_dict', 'vader_compound']].head(5))
```

```python
from collections import Counter
from itertools import tee

def get_ngrams(words, n=2):
    words, temp = tee(words)
    return zip(*[temp] * n)

bigrams = Counter(get_ngrams(token_df['word']))
print(bigrams.most_common(10))  # Show most frequent word pairs
```
